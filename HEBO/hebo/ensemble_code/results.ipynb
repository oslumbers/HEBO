```python
import pickle
import pandas as pd
import numpy as np
import os
```

```python
datasets = ['digits', 'wine']
num_ens = [1,2,3]

results = {}

for dataset in datasets:
    results[dataset] = {}
    for ens in num_ens:
        results[dataset][ens] = {}
        with open(f'/home/oslumbers/Documents/git_repos/HEBO/HEBO/hebo/ensemble_code/results_testing/random_forest/{dataset}/num_ens_{ens}.p', 'rb') as f:
            data = pickle.load(f)

        seeds = []
        for i in range(len(data)):
            bests = []
            best = data[i][0]['metric'].values[0]
            bests.append(best)
            for j in range(1,len(data[i][0]['metric'].values)):
                val = data[i][0]['metric'].values[j]
                if val < best:
                    best = val
                bests.append(best)
            seeds.append(bests)
        means = np.mean(seeds, axis=0)
        stds = np.std(seeds, axis=0)
        results[dataset][ens]['iter_mean'] = means
        results[dataset][ens]['iter_std'] = stds


for dataset in datasets:
    final_values = []
    data = results[dataset]
    for key in data.keys():
        final_mean = data[key]['iter_mean'][-1]
        final_std = data[key]['iter_std'][-1]
        final_values.append((key, final_mean, final_std))

    # Create a pandas DataFrame
    df = pd.DataFrame(final_values, columns=['Ensemble Size', 'Final Mean', 'Final Std'])
    csv_dir = f'/home/oslumbers/Documents/git_repos/HEBO/HEBO/hebo/ensemble_code/results_testing/random_forest/csvs/{dataset}'
    if not os.path.exists(csv_dir):
        os.makedirs(csv_dir)

    reference_mean = df.loc[df['Ensemble Size'] == 1, 'Final Mean'].values[0]
    df['Percentage Improvement'] = (reference_mean - df['Final Mean']) / reference_mean * 100


    df.to_csv(os.path.join(csv_dir, f'{dataset}.csv'), index=False)
```
